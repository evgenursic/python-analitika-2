```python
def feature_engineering(data, non_numeric_columns=None):
    for column in non_numeric_columns:
        le = LabelEncoder()
        data[column] = le.fit_transform(data[column])
    return data    
```

```python
def top10_subplot(volatility_series, title):
    fig, axes = plt.subplots(nrows=1, ncols=2, figsize=(10, 6))
    
    ax = (volatility_series[:10].plot.bar(color='darkred', ax=axes[0]))

    fig.suptitle(title)

    ax.set_ylabel('% change')

    ax = (volatility_series[-10:].plot.bar(color='darkblue', ax=axes[1]))
    

    return fig, ax

DTITLE = "24 hours top losers and winners"
 
fig, ax = top10_subplot(volatility['percent_change_24h'] ,DTITLE)
plt.show()
```

### Handling the missing values (part ii)

```python
# 12 manjkajočih vrednosti želimo zapolniti s povprečno vrednostjo ostalih
```

```python
cc_apps.fillna(cc_apps.mean(), inplace=True)
```

```python
cc_apps.isnull().sum()
```

### Handling the missing values (part iii)

```python
for col in cc_apps: # samo stolpci, ki so stringi
    if cc_apps[col].dtypes == 'object':
        cc_apps = cc_apps.fillna(cc_apps[col].value_counts().index[0])
```

```python
cc_apps.isnull().sum()
```

### Preprocessing the data (part i)

```python
from sklearn.preprocessing import LabelEncoder # nenumerične pretvoriti razdeliti podatke na train in test in feturje0 in 1
```

```python
le = LabelEncoder()
```

```python
for col in cc_apps:
    if cc_apps[col].dtypes == 'object':
        cc_apps[col]  =le.fit_transform(cc_apps[col])
```

```python
cc_apps.head()
```

### Splitting the dataset into train and test sets

```python
from sklearn.model_selection import train_test_split

cc_apps = cc_apps.drop([11,13], axis=1)
cc_apps = cc_apps.values
```

```python
X = cc_apps[:, 0:13]
y = cc_apps[:, 13]
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33, random_state=42)
```

### Preprocessing the data (part ii)

```python
from sklearn.preprocessing import MinMaxScaler


scaler = MinMaxScaler(feature_range=(0,1))

rescaled_train = scaler.fit_transform(X_train)
rescaled_test = scaler.fit_transform(X_test)
```

### Fitting a logistic regression model to the train set

```python
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression()

logreg.fit(rescaledX_train, y_train)

```

### Making predictions and evaluating performance
